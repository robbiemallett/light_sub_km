```python
df = pd.read_csv('hs7.csv',index_col='mean_depth')

def get_hs7(mean_depth):
    
    mean_depth_2dp = np.round(mean_depth+1e-6,2)
    
    hs7 = np.array(df.loc[mean_depth_2dp])
    
    return hs7
    
# Here's how to get seven snow depths of equal probability
# from one mean depth value

get_hs7(0.1285)
```
